# Convert 0 and 1 values of senior citizen to yes/no to make it to easier to understand

```python
def conv(value):
  if value== 1:
    return "yes"
  else:
    return "no"

df['SeniorCitizen'] = df["SeniorCitizen"].apply(conv)
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x= 'gender',data = df, hue = 'Churn')
plt.title("churn by gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x= 'SeniorCitizen',data = df, hue = 'Churn')
plt.title("churn by SeniorCitizen")
plt.show()
```

```python
counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])
percentages = counts.div(counts.sum(axis=1), axis=0) * 100


plt.figure(figsize=(4,4))
bottom_val = [0] * len(percentages)

for churn_status in percentages.columns:
    plt.bar(percentages.index, percentages[churn_status], bottom=bottom_val, label=churn_status)
    bottom_val = [i + j for i, j in zip(bottom_val, percentages[churn_status])]


for i, (idx, row) in enumerate(percentages.iterrows()):
    cumulative = 0
    for churn_status in percentages.columns:
        height = row[churn_status]
        plt.text(idx, cumulative + height/2, f'{height:.1f}%', ha='center', va='center', fontsize=9)
        cumulative += height


plt.xticks([0, 1], ['Not Senior', 'Senior'])
plt.title('Churn by SeniorCitizen (Stacked %)')
plt.ylabel('Percentage')
plt.legend(title='Churn', bbox_to_anchor=(0.9,0.9))
plt.tight_layout()
plt.show()
```
